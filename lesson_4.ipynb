### Задача 1
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].  
Найдите ее среднее значение и дисперсию.

```python
# среднее значение
M = (800 - 200)/2
# дисперси
D = pow(800 - 200, 2) / 12
print(f'Среднее значение {M} дисперсия {D}')
```

### Задача 4
Рост взрослого населения города X имеет нормальное распределение.  
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.  
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:  
1. больше 182 см
1. больше 190 см
1. от 166 см до 190 см
1. от 166 см до 182 см
1. от 158 см до 190 см
1. не выше 150 см или не ниже 190 см
1. не выше 150 см или не ниже 198 см
1. ниже 166 см.

```python
def get_z_score(x, Mu = 174, std = 8):
    return (x - Mu)/std
```

```python
from scipy.stats import norm

# 1 больше 182 см
z_score1 = get_z_score(182)
print(f'1. Рост 182 см лежит в {z_score1} сигм/е от среднеарифметического значения')

P1 = 1 - norm.cdf(z_score1)
print(f'   Вероятность выбрать человека с ростом выше 182 равна {P1:.5f}\n')

# 2 больше 190 см
z_score2 = get_z_score(190)
print(f'2. Рост 190 см лежит в {z_score2} сигм/е от среднеарифметического значения')

P2 = 1 - norm.cdf(z_score2)
print(f'   Вероятность выбрать человека с ростом выше 190 равна {P2:.5f}\n')

# 3 от 166 см до 190 см
z_score3_1 = get_z_score(166)
z_score3_2 = get_z_score(190)
print(f'3. Рост от 166 см до 190 см лежит между {z_score3_1} и {z_score3_2} сигамами')

P3 = norm.cdf(z_score3_2) - norm.cdf(z_score3_1)
print(f'   Вероятность выбрать человека с ростом от 166 см до 190 равна {P3:.5f}\n')

# 4 от 166 см до 182 см
z_score4_1 = get_z_score(166)
z_score4_2 = get_z_score(182)
print(f'4. Рост от 166 см до 182 см лежит между {z_score4_1} и {z_score4_2} сигамами')

P4 = norm.cdf(z_score4_2) - norm.cdf(z_score4_1)
print(f'   Вероятность выбрать человека с ростом от 166 см до 182 равна {P4:.5f}\n')

# 5 от 158 см до 190 см
z_score5_1 = get_z_score(158)
z_score5_2 = get_z_score(190)
print(f'5. Рост от 158 см до 190 см лежит между {z_score5_1} и {z_score5_2} сигамами')

P5 = norm.cdf(z_score5_2) - norm.cdf(z_score5_1)
print(f'   Вероятность выбрать человека с ростом от 158 см до 190 равна {P5:.5f}\n')

# 6 не выше 150 см или не ниже 190 см
z_score6_1 = get_z_score(150)
z_score6_2 = get_z_score(190)
print(f'6. Рост от 150 см до 190 см лежит вне диапазона от {z_score6_1} до {z_score6_2} сигамами')

P6 = norm.cdf(z_score6_1) + (1 - norm.cdf(z_score6_2))
print(f'   Вероятность выбрать человека с ростом не выше 150 см или не ниже 190 см равна {P6:.5f}\n')

# 7 не выше 150 см или не ниже 198 см
z_score7_1 = get_z_score(150)
z_score7_2 = get_z_score(198)
print(f'7. Рост от 150 см до 198 см лежит вне диапазона от {z_score7_1} до {z_score7_2} сигамами')

P7 = norm.cdf(z_score7_1) + (1 - norm.cdf(z_score7_2))
print(f'   Вероятность выбрать человека с ростом не выше 150 см или не ниже 198 см равна {P7:.5f}\n')

# 8 ниже 166 см.
z_score8 = get_z_score(166)
print(f'8. Рост 166 см лежит в {z_score1} сигм/е от среднеарифметического значения')

P8 = norm.cdf(z_score8)
print(f'   Вероятность выбрать человека с ростом ниже 166 равна {P8:.5f}\n')

```
